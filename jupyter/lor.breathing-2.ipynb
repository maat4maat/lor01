### Тест на нормальность

```r
for (group_name in levels(lorl[[groupping_variable]])) {
    display(paste0('======================== ', group_name, ' ========================'))
    tmpset = lorl %>%
        filter(группа == group_name)

    for (i in 1:nlevels(tmpset$"время")) {
        display(paste0("Время — ", levels(tmpset$"время")[i]))
        ss = subset(tmpset, tmpset$"время" == levels(tmpset$"время")[i])
        display(try(print(shapiro.test(ss[[parname]]))))
        
        display(try(bartlett.test(tmpset[[parname]] ~ tmpset$"время")))
        display(paste0(''))
    }
    display(paste0(''))
}
```

### Сравнение, нормальные распределения

```r
for (group_name in levels(lorl[[groupping_variable]])) {
    cat(paste0('\n======================== ', group_name, ' ========================\n'))
    tmpset = lorl %>%
        filter(группа == group_name)

    if (nlevels(parameter) < 3) {
        display(t.test(tmpset[[parname]] ~ tmpset$"время", paired = TRUE))
    } else {
        agg1.aov = aov(tmpset[[parname]] ~ tmpset$"время")
        print(agg1.aov)
        print(summary(agg1.aov))
        cat("\nTukey post hoc test\n")
        print(TukeyHSD(agg1.aov, conf.level=.95))
        op=par(mar=c(4,9,4,2), lwd=2)
        plot(TukeyHSD(agg1.aov, conf.level=.95), las=1, col="brown", cex.axis=1, line.lwd=2)
        par(op)
    }    
}
```

### Сравнение, распределение не нормальное

```r
for (group_name in levels(lorl[[groupping_variable]])) {
    cat(paste0('\n\n======================== ', group_name, ' ========================\n'))
    tmpset = lorl %>%
        filter(группа == group_name)

    if (nlevels(parameter) < 3) {
        display(wilcox.test(tmpset[[parname]] ~ tmpset$"время", paired = TRUE))
        print(independence_test(tmpset[[parname]] ~ tmpset$"время", data = tmpset,
                          alternative = "two.sided",
                          ytrafo = function(data) trafo(data, numeric_trafo = rank_trafo),
                          xtrafo = function(data) trafo(data, ordered_trafo = ff)
                          )
              )
    } else {
        print(kruskal.test(tmpset[[parname]] ~ tmpset[["время"]]))
        print(kruskalmc(tmpset[[parname]] ~ tmpset[["время"]]))
    }
}
```

### Сравнение, нормальные распределения

```r
for (group_name in levels(lorl[[groupping_variable]])) {
    cat(paste0('\n======================== ', group_name, ' ========================\n'))
    tmpset = lorl %>%
        filter(группа == group_name)

    if (nlevels(parameter) < 3) {
        display(t.test(tmpset[[parname]] ~ tmpset$"время", paired = TRUE))
    } else {
        agg1.aov = aov(tmpset[[parname]] ~ tmpset$"время")
        print(agg1.aov)
        print(summary(agg1.aov))
        cat("\nTukey post hoc test\n")
        print(TukeyHSD(agg1.aov, conf.level=.95))
        op=par(mar=c(4,9,4,2), lwd=2)
        plot(TukeyHSD(agg1.aov, conf.level=.95), las=1, col="brown", cex.axis=1, line.lwd=2)
        par(op)
    }    
}
```

### Сравнение, распределение не нормальное

```r
for (group_name in levels(lorl[[groupping_variable]])) {
    cat(paste0('\n\n======================== ', group_name, ' ========================\n'))
    tmpset = lorl %>%
        filter(группа == group_name)

    if (nlevels(parameter) < 3) {
        display(wilcox.test(tmpset[[parname]] ~ tmpset$"время", paired = TRUE))
        print(independence_test(tmpset[[parname]] ~ tmpset$"время", data = tmpset,
                          alternative = "two.sided",
                          ytrafo = function(data) trafo(data, numeric_trafo = rank_trafo),
                          xtrafo = function(data) trafo(data, ordered_trafo = ff)
                          )
              )
    } else {
        print(kruskal.test(tmpset[[parname]] ~ tmpset[["время"]]))
        print(kruskalmc(tmpset[[parname]] ~ tmpset[["время"]]))
    }
}
```
