### Post-hoc анализ

A significant two-way interaction indicates that the impact that one factor (e.g., group) has on the outcome variable (e.g., OAK) depends on the level of the other factor (e.g., time) (and vice versa). So, you can decompose a significant two-way interaction into:
- Simple main effect: run one-way model of the first variable (factor A) at each level of the second variable (factor B),
- Simple pairwise comparisons: if the simple main effect is significant, run multiple pairwise comparisons to determine which groups are different.

For a non-significant two-way interaction, you need to determine whether you have any statistically significant main effects from the ANOVA output.

```r
options(dplyr.width = Inf)
for (i in match("ОАК.эритроциты", names(lorl)):match("ОАК.СОЭ", names(lorl))) {
  parname <- colnames(lorl)[i]

  ss <- lorl %>%
    select("uid", "группа", "пол", "time", parname) %>%
    filter(time != "до СК")
  levels(ss$time)[2] <- NA
  ss <- cbind(ss, logparname = log(ss[[parname]]))

  cat("\n\n=== Параметр:", parname, "\n")

  # Effect of treatment at each time point
  cat("    Эффект группы в каждой точке времени\n")
  cat("        Эффект группы в каждой точке времени\n")
  one.way <- ss %>%
    group_by(time) %>%
    anova_test(dv = parname, wid = uid, between = группа) %>%
    get_anova_table() %>%
    adjust_pvalue(method = "bonferroni")
  print(one.way)

  # Pairwise comparisons between treatment groups
  cat("\n        Сравнение в группах\n")
  fm <- formula(paste0(parname, " ~ группа"))
  pwc <- ss %>%
    group_by(time) %>%
    pairwise_t_test(
      fm,
      paired = FALSE,
      p.adjust.method = "bonferroni"
    )
  print(pwc)

  # Effect of time
  cat("\n    Эффект группы\n")
  cat("        Эффект времени для каждой группы\n")
  one.way <- ss %>%
    group_by(группа) %>%
    anova_test(dv = parname, wid = uid, within = time) %>%
    get_anova_table() %>%
    adjust_pvalue(method = "bonferroni")
  print(one.way)

  # Pairwise comparisons between time points
  cat("\n        Сравнение между моментами времени\n")
  fm <- formula(paste0(parname, " ~ time"))
  pwc <- ss %>%
    group_by(группа) %>%
    pairwise_t_test(
      fm,
      paired = TRUE,
      p.adjust.method = "bonferroni"
    )
  print(pwc)
}
```

### Post-hoc анализ

A significant two-way interaction indicates that the impact that one factor (e.g., group) has on the outcome variable (e.g., OAK) depends on the level of the other factor (e.g., time) (and vice versa). So, you can decompose a significant two-way interaction into:
- Simple main effect: run one-way model of the first variable (factor A) at each level of the second variable (factor B),
- Simple pairwise comparisons: if the simple main effect is significant, run multiple pairwise comparisons to determine which groups are different.

For a non-significant two-way interaction, you need to determine whether you have any statistically significant main effects from the ANOVA output.

```r
options(dplyr.width = Inf)
for (i in match("ОАК.эритроциты", names(lorl.norm)):match("ОАК.СОЭ", names(lorl.norm))) {
  parname <- colnames(lorl)[i]

  ss <- lorl.norm %>%
    select("uid", "группа", "пол", "time", parname) %>%
    filter(time != "до СК")
  levels(ss$time)[2] <- NA
  ss <- cbind(ss, logparname = log(ss[[parname]]))

  cat("\n\n=== Параметр:", parname, "\n")

  # Effect of treatment at each time point
  cat("    Эффект группы в каждой точке времени\n")
  cat("        Эффект группы в каждой точке времени\n")
  one.way <- try(ss %>%
    group_by(time) %>%
    anova_test(dv = parname, wid = uid, between = группа) %>%
    get_anova_table() %>%
    adjust_pvalue(method = "bonferroni"))
  print(one.way)

  # Pairwise comparisons between treatment groups
  cat("\n        Сравнение в группах\n")
  fm <- formula(paste0(parname, " ~ группа"))
  pwc <- ss %>%
    group_by(time) %>%
    pairwise_t_test(
      fm,
      paired = FALSE,
      p.adjust.method = "bonferroni"
    )
  print(pwc)

  # Effect of time
  cat("\n    Эффект группы\n")
  cat("        Эффект времени для каждой группы\n")
  one.way <- try(ss %>%
    group_by(группа) %>%
    anova_test(dv = parname, wid = uid, within = time) %>%
    get_anova_table() %>%
    adjust_pvalue(method = "bonferroni"))
  print(one.way)

  # Pairwise comparisons between time points
  cat("\n        Сравнение между моментами времени\n")
  fm <- formula(paste0(parname, " ~ time"))
  pwc <- ss %>%
    group_by(группа) %>%
    pairwise_t_test(
      fm,
      paired = TRUE,
      p.adjust.method = "bonferroni"
    )
  print(pwc)
}
```

```r
### Непараметрические тесты
```

If the data do not meet the assumptions of normality, consider non-parametric tests like the Friedman test for within-group comparisons and Kruskal-Wallis test for between-group comparisons.

```r
# Friedman test for within-group comparisons
ss.friedman <- friedman.test(ОАК.эритроциты ~ time | uid, data = ss)
print(ss.friedman)

# Kruskal-Wallis test for between-group comparisons at each time point
ss.kruskal <- kruskal.test(ОАК.эритроциты ~ группа, data = subset(ss, time == "1 год"))
print(ss.kruskal)
```
